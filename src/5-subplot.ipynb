### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
sizes = np.random.rand(50) * 100  # Bubble sizes for scatter plot

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line chart in the first subplot
axs[0, 0].plot(x, y, label="sin(x)")
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend()

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, s=sizes, alpha=0.6, color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
